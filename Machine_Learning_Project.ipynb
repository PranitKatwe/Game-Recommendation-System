```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
data = pd.read_csv('games-features.csv')
data.head()
```

```python
data.info()
```

```python
data.replace(0, np.nan, inplace=True)
data.replace(' ', np.nan, inplace=True)
data
```

```python
missingdata = data.isnull().sum().sort_values(ascending=False).head(40)
missingdata
```

```python
plt.figure(figsize=(15, 12))
sns.barplot(x=missingdata, y=missingdata.index, palette='viridis')
plt.title('Top 40 Columns with Missing Data')
plt.xlabel('Number of Missing Values')
plt.ylabel('Column Name')
plt.show()
```

```python
# For now Dropping Unnecessary Columns
# More columns can be dropped depending on the what should be worked on
drop_cols = ['ResponseID','ResponseName','DRMNotice', 'ExtUserAcctNotice', 'DemoCount','LinuxRecReqsText','MacRecReqsText','MacMinReqsText','MovieCount','PackageCount','SteamSpyOwnersVariance','SteamSpyPlayersVariance','SupportEmail','SupportURL','Website']

data.drop(drop_cols, axis=1, inplace=True)
```

```python
data
```

```python
data.info()
```

```python
# Visualization for Data Description
category_boolean_columns = ['CategorySinglePlayer', 'CategoryMultiplayer', 'CategoryCoop',
                             'CategoryMMO', 'CategoryInAppPurchase', 'CategoryIncludeSrcSDK',
                             'CategoryIncludeLevelEditor', 'CategoryVRSupport']


fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(8, 8))
fig.suptitle('Distribution of Category Columns', fontsize=16)


axes = axes.flatten()

for i, column in enumerate(category_boolean_columns):
    sns.countplot(x=column, data=data, ax=axes[i], palette='Set2')
    axes[i].set_title(column)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    
plt.tight_layout()
plt.show()
```

```python
game_recommend = data.groupby('QueryName')['RecommendationCount'].sum().sort_values(ascending=False).head(10)
game_recommend
```

```python
# Pie Chart for Top 10 Recommended Games
labels = game_recommend.index
sizes = game_recommend.values

plt.figure(figsize=(15, 10))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 10 Most Recommended Games')
plt.show()
```

```python
data.replace(np.nan, 0 ,inplace=True)
data.replace(np.nan,' ',inplace=True)

data['RequiredAge'] = data['RequiredAge'].astype(int)
data['DeveloperCount'] = data['DeveloperCount'].astype(int)
data['DLCCount'] = data['DLCCount'].astype(int)
data['Metacritic'] = data['Metacritic'].astype(int)
data['RecommendationCount'] = data['RecommendationCount'].astype(int)
data['PublisherCount'] = data['PublisherCount'].astype(int)
data['ScreenshotCount'] = data['ScreenshotCount'].astype(int)
data['SteamSpyOwners'] = data['SteamSpyOwners'].astype(int)
data['SteamSpyPlayersEstimate'] = data['SteamSpyPlayersEstimate'].astype(int)
data['AchievementCount'] = data['AchievementCount'].astype(int)
data['AchievementHighlightedCount'] = data['AchievementHighlightedCount'].astype(int)
```

```python
data
```

```python
data.info()
```

```python
plt.figure(figsize=(10, 6))
rating = data[data['Metacritic'] != 0]
sns.histplot(rating['Metacritic'], bins=30, kde=True)
plt.title('Distribution of Metacritic Scores')
plt.xlabel('Metacritic Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Understanding Correlation for further Machine learning algporithms feature selection
correlation_matrix = data.corr()
plt.figure(figsize=(50, 50))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
data['IsFree'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, labels=['Not Free', 'Free'])
plt.title('Percentage of Free Games')
plt.show()
```

```python
# How Indie Games(Developed by less than 10 developers without any budget) are related to Metacritic
import plotly.express as px

fig = px.scatter(data, x='SteamSpyOwners', y='RecommendationCount', color='GenreIsIndie', hover_data=['QueryName'],
                 title='Interactive Scatter Plot - Indie Games vs Recommendation Count',
                 labels={'SteamSpyOwners': 'Owners', 'RecommendationCount': 'Recommendation Count'})
fig.show()
```
