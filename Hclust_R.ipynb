```r
df <- data.frame(game)
df
```

```r
game_name <- df$QueryName
game_name
```

```r
df <- df[,-c(1)]
df
```

```r
normalized_data <- as.data.frame(scale(df))
```

```r
cosin <- stats::dist(df, method = "manhattan")
```

```r
options(repr.plot.width=20, repr.plot.height=15)
hclust_cosin <- hclust(cosin, method = "ward.D2")
plot(hclust_cosin, cex = 0.9, hang = -14, main = "Dendrogram of Cosine Distance", labels = game_name)
rect.hclust(hclust_cosin, k = 4, border = "red")
```
