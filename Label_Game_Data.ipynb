```python
# Calculate mean
mean_steam_spy_owners = data['SteamSpyOwners'].mean()
mean_steam_spy_players_estimate = data['SteamSpyPlayersEstimate'].mean()
mean_recommendation_count = data['RecommendationCount'].mean()


# Calculate median
median_steam_spy_owners = data['SteamSpyOwners'].median()
median_steam_spy_players_estimate = data['SteamSpyPlayersEstimate'].median()
median_recommendation_count = data['RecommendationCount'].median()

print("Mean SteamSpyOwners:", mean_steam_spy_owners)
print("Median SteamSpyOwners:", median_steam_spy_owners)

print("Mean SteamSpyPlayersEstimate:", mean_steam_spy_players_estimate)
print("Median SteamSpyPlayersEstimate:", median_steam_spy_players_estimate)

print("Mean RecommendationCount:", mean_recommendation_count)
print("Median RecommendationCount:", median_recommendation_count)
```

```python
metacritic_threshold = 80
recommendation_count_threshold = 1000
steam_spy_owners_threshold = 150000
steam_spy_players_estimate_threshold = 100000

def create_label(row):
    if (row['Metacritic'] > metacritic_threshold) or \
       (row['RecommendationCount'] > recommendation_count_threshold) and \
       (row['SteamSpyOwners'] > steam_spy_owners_threshold) and \
       (row['SteamSpyPlayersEstimate'] > steam_spy_players_estimate_threshold):
        return 1  # Recommended
    else:
        return 0  # Not Recommended

data['Recommended'] = data.apply(create_label, axis=1)
```
